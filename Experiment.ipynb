```python
import numpy as np

# Create a NumPy array
array = np.array([1, 2, 3, 4, 5])

# Perform basic operations
array_squared = array ** 2
array_sum = np.sum(array)

# Print the results
print("Original Array:", array)
print("Squared Array:", array_squared)
print("Sum of Array Elements:", array_sum)
```

```python
import numpy as np

# Create two 2D NumPy arrays (matrices)
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication
matrix_product = np.dot(matrix_a, matrix_b)

# Calculate the transpose of matrix_a
matrix_transpose = np.transpose(matrix_a)

# Print the results
print("Matrix A:\n", matrix_a)
print("Matrix B:\n", matrix_b)
print("Matrix Product (A * B):\n", matrix_product)
print("Transpose of Matrix A:\n", matrix_transpose)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine of each x value

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sine Wave', color='blue')

# Add title and labels
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # Add x-axis
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Add y-axis
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)  # Add grid
plt.legend()  # Show legend

# Show the plot
plt.show()
```
